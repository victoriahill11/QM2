```python
import numpy as np
import statsmodels.formula.api as smf

# Generate some sample data
x1 = np.random.rand(100)
x2 = np.random.rand(100)
y = 2 + 3*x1 + 3*x2 + np.random.randn(100)

# Create a DataFrame
df = pd.DataFrame({'x1': x1, 'x2': x2, 'y': y})

# Fit the linear regression model with a same slope assumption
model = smf.ols('y ~ x1 + x2 - 1', data=df).fit()

# Print the model summary
print(model.summary())
```

```python
import numpy as np  # Import numpy

# Create a dictionary to store the regression models
models = {}

# Fit a linear regression model with a same slope assumption for each country
for Country in df['Country'].unique():
    country_data = df[df['Country'] == country]
    # Fix: Use the correct column name 'Annual Food price index'
    model = smf.ols('Q("Annual Food price index") ~ Year - 1', data=country_data).fit()
    models[country] = model

# Create a figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data points for each country
for country in df['Country'].unique():
    country_data = df[df['Country'] == country]
    # Fix: Use the correct column name 'Annual Food price index'
    ax.scatter(country_data['Year'], country_data['Annual Food price index'], label=country)

# Plot the regression lines for each country
for country, model in models.items():
    x = np.linspace(df['Year'].min(), df['Year'].max(), 100)
    y = model.params['Year'] * x
    ax.plot(x, y, label=f'{country} (slope={model.params["Year"]:.2f})')

# Add labels and title
ax.set_xlabel('Year')
ax.set_ylabel('Food Price Index')
ax.set_title('Linear Regression with Same Slope for Multiple Countries')

# Add legend
ax.legend()

# Show the plot
plt.show()

# Create an empty list to store the data for the Altair chart
altair_data = []

# Add the original data points to the list
for country in df['Country'].unique():
    country_data = df[df['Country'] == country]
    for _, row in country_data.iterrows():
        # Fix: Use the correct column name 'Annual Food price index'
        altair_data.append({'Country': country, 'Year': row['Year'], 'FoodPriceIndex': row['Annual Food price index'], 'Type': 'Original Data'})

# Add the regression lines to the list
for country, model in models.items():
    x = np.linspace(df['Year'].min(), df['Year'].max(), 100)
    y = model.params['Year'] * x
    for i in range(len(x)):
        altair_data.append({'Country': country, 'Year': x[i], 'FoodPriceIndex': y[i], 'Type': 'Regression Line'})

# Create an Altair chart
chart = alt.Chart(pd.DataFrame(altair_data)).mark_line(point=True).encode(
    x='Year',
    y='FoodPriceIndex',
    color='Country',
    tooltip=['Country', 'Year', 'FoodPriceIndex'],
    shape='Type'
).properties(
    title='Linear Regression with Same Slope for Multiple Countries'
).interactive()

# Save the chart
chart.save('linear_regression_same_slope_altair.json')
```
